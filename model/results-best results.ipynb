```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from utils import load_json
```

```python
def get_best_results(json_fname):
    results = load_json(json_fname)
    best_results = []
    super_keys = []
    for missing_percent in range(5, 25, 5):
        missing_percent = str(missing_percent)
        df = pd.DataFrame(results[missing_percent])
        df['sub_auc'] = 1 - df.auc
        df['score'] = df.nrmse + df.error + df.sub_auc
        for fold in range(5):
            im = df.loc[df.fold == fold].score.idxmin()
            best_result = dict(df.loc[im])
            best_result.update({'missing_percent': missing_percent,
                               'fold': fold})
            best_results.append(best_result)

    best_results = pd.DataFrame(best_results)
    best_results.missing_percent = best_results.missing_percent.astype('int')
    return best_results
```

```python
churn = get_best_results('./results/1119churn_DSAN.json')
churn.groupby('missing_percent').mean().round(4)
```

```python
churn.groupby('missing_percent').mean().mean(axis=0)
```
